4. –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ, –¥–ª—è —Å–∫–æ–ª—å–∫–∏—Ö —Ä–µ—Ü–µ–ø—Ç–æ–≤ —É–∫–∞–∑–∞–Ω–∞ –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å, —Ä–∞–≤–Ω–∞—è –Ω—É–ª—é. –ó–∞–º–µ–Ω–∏—Ç–µ –¥–ª—è —Ç–∞–∫–∏—Ö —Å—Ç—Ä–æ–∫ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ –¥–∞–Ω–Ω–æ–º —Å—Ç–æ–ª–±—Ü–µ –Ω–∞ 1.

5. –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ, —Å–∫–æ–ª—å–∫–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Ä–µ—Ü–µ–ø—Ç–æ–≤ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ.

6. –°–∫–æ–ª—å–∫–æ –∏ –∫–∞–∫–∏—Ö —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∫–æ–ª-–≤–∞ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤ –ø—Ä–∏—Å—É—Ç–≤—É–µ—Ç –≤ —Ä–µ—Ü–µ–ø—Ç–∞—Ö –∏–∑ –¥–∞—Ç–∞—Å–µ—Ç–∞?

7. –°–æ–∑–¥–∞–π—Ç–µ –≤–µ—Ä—Å–∏—é –º–∞—Å—Å–∏–≤–∞, —Å–æ–¥–µ—Ä–∂–∞—â—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é —Ç–æ–ª—å–∫–æ –æ —Ä–µ—Ü–µ–ø—Ç–∞—Ö, —Å–æ—Å—Ç–æ—è—â–∏—Ö –Ω–µ –±–æ–ª–µ–µ —á–µ–º –∏–∑ 5 –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤.

8. –î–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–µ—Ü–µ–ø—Ç–∞ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ, —Å–∫–æ–ª—å–∫–æ –≤ —Å—Ä–µ–¥–Ω–µ–º –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤ –ø—Ä–∏—Ö–æ–¥–∏—Ç—Å—è –Ω–∞ –æ–¥–Ω—É –º–∏–Ω—É—Ç—É —Ä–µ—Ü–µ–ø—Ç–∞. –ù–∞–π–¥–∏—Ç–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —ç—Ç–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –¥–ª—è –≤—Å–µ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞

9. –í—ã—á–∏—Å–ª–∏—Ç–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤ –¥–ª—è —Ç–æ–ø-100 —Ä–µ—Ü–µ–ø—Ç–æ–≤ —Å –Ω–∞–∏–±–æ–ª—å—à–µ–π –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é

10. –í—ã–±–µ—Ä–∏—Ç–µ —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º –∏ –≤—ã–≤–µ–¥–∏—Ç–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ 10 —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ä–µ—Ü–µ–ø—Ç–∞—Ö

11. –í—ã–≤–µ–¥–∏—Ç–µ –ø—Ä–æ—Ü–µ–Ω—Ç —Ä–µ—Ü–µ–ø—Ç–æ–≤, –∫–æ–ª-–≤–æ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤ –≤ –∫–æ—Ç–æ—Ä—ã—Ö –º–µ–Ω—å—à–µ —Å—Ä–µ–¥–Ω–µ–≥–æ.

12. –ù–∞–∑–æ–≤–µ–º "–ø—Ä–æ—Å—Ç—ã–º" —Ç–∞–∫–æ–π —Ä–µ—Ü–µ–ø—Ç, –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∫–æ—Ç–æ—Ä–æ–≥–æ –Ω–µ –±–æ–ª—å—à–µ 20 –º–∏–Ω—É—Ç –∏ –∫–æ–ª-–≤–æ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤ –≤ –∫–æ—Ç–æ—Ä–æ–º –Ω–µ –±–æ–ª—å—à–µ 5. –°–æ–∑–¥–∞–π—Ç–µ –≤–µ—Ä—Å–∏—é –¥–∞—Ç–∞—Å–µ—Ç–∞ —Å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–º —Å—Ç–æ–ª–±—Ü–æ–º, –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –∫–æ—Ç–æ—Ä–æ–≥–æ —è–≤–ª—è—é—Ç—Å—è 1, –µ—Å–ª–∏ —Ä–µ—Ü–µ–ø—Ç –ø—Ä–æ—Å—Ç–æ–π, –∏ 0 –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

13. –í—ã–≤–µ–¥–∏—Ç–µ –ø—Ä–æ—Ü–µ–Ω—Ç "–ø—Ä–æ—Å—Ç—ã—Ö" —Ä–µ—Ü–µ–ø—Ç–æ–≤ –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ

14. –†–∞–∑–¥–µ–ª–∏–º —Ä–µ—Ü–µ–ø—Ç—ã –Ω–∞ –≥—Ä—É–ø–ø—ã –ø–æ —Å–ª–µ–¥—É—é—â–µ–º—É –ø—Ä–∞–≤–∏–ª—É. –ù–∞–∑–æ–≤–µ–º —Ä–µ—Ü–µ–ø—Ç—ã –∫–æ—Ä–æ—Ç–∫–∏–º–∏, –µ—Å–ª–∏ –∏—Ö –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –º–µ–Ω–µ–µ 10 –º–∏–Ω—É—Ç; —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º–∏, –µ—Å–ª–∏ –∏—Ö –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –±–æ–ª–µ–µ 10, –Ω–æ –º–µ–Ω–µ–µ 20 –º–∏–Ω—É—Ç; –∏ –¥–ª–∏–Ω–Ω—ã–º–∏, –µ—Å–ª–∏ –∏—Ö –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –Ω–µ –º–µ–Ω–µ–µ 20 –º–∏–Ω—É—Ç. –°–æ–∑–¥–∞–π—Ç–µ —Ç—Ä–µ—Ö–º–µ—Ä–Ω—ã–π –º–∞—Å—Å–∏–≤, –≥–¥–µ –Ω—É–ª–µ–≤–∞—è –æ—Å—å –æ—Ç–≤–µ—á–∞–µ—Ç –∑–∞ –Ω–æ–º–µ—Ä –≥—Ä—É–ø–ø—ã (–∫–æ—Ä–æ—Ç–∫–∏–π, —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–π –∏–ª–∏ –¥–ª–∏–Ω–Ω—ã–π —Ä–µ—Ü–µ–ø—Ç), –ø–µ—Ä–≤–∞—è –æ—Å—å - –∑–∞ —Å–∞–º —Ä–µ—Ü–µ–ø—Ç –∏ –≤—Ç–æ—Ä–∞—è –æ—Å—å - –∑–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ —Ä–µ—Ü–µ–ø—Ç–∞. –í—ã–±–µ—Ä–∏—Ç–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–µ—Ü–µ–ø—Ç–æ–≤ –∏–∑ –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø—ã —Ç–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —á—Ç–æ–±—ã –±—ã–ª–æ –≤–æ–∑–º–æ–∂–Ω–æ —Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞—Ç—å —Ç—Ä–µ—Ö–º–µ—Ä–Ω—ã–π –º–∞—Å—Å–∏–≤. –í—ã–≤–µ–¥–∏—Ç–µ —Ñ–æ—Ä–º—É –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞.

```python
# ‚Ññ1–§–∞–π–ª minutes_n_ingredients.csv —Å–æ–¥–µ—Ä–∂–∏—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ–± –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä–µ —Ä–µ—Ü–µ–ø—Ç–∞, 
#–≤—Ä–µ–º–µ–Ω–∏ –µ–≥–æ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –≤ –º–∏–Ω—É—Ç–∞—Ö –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤. 
#–°—á–∏—Ç–∞–π—Ç–µ –¥–∞–Ω–Ω—ã–µ –∏–∑ —ç—Ç–æ–≥–æ —Ñ–∞–π–ª–∞ –≤ –≤–∏–¥–µ –º–∞—Å—Å–∏–≤–∞ numpy —Ç–∏–ø–∞ int32, –∏—Å–ø–æ–ª—å–∑—É—è np.loadtxt. 
#–í—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ —ç–∫—Ä–∞–Ω –ø–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫ –º–∞—Å—Å–∏–≤–∞.
import numpy as np
data=np.loadtxt('minutes_n_ingredients.csv', dtype=np.int32,delimiter=',')
print(data[:5])
```

```python
# ‚Ññ2 –í—ã—á–∏—Å–ª–∏—Ç–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, –º–∏–Ω–∏–º—É–º, –º–∞–∫—Å–∏–º—É–º –∏ –º–µ–¥–∏–∞–Ω—É –ø–æ –∫–∞–∂–¥–æ–º—É –∏–∑ —Å—Ç–æ–ª–±—Ü–æ–≤, –∫—Ä–æ–º–µ –ø–µ—Ä–≤–æ–≥–æ.
import numpy as np
data=np.loadtxt('minutes_n_ingredients.csv', dtype=np.int32,delimiter=',')
data2=data[:,1:]
sred=np.mean(data2)
mini=np.min(data2)
maksi=np.max(data2)
med=np.median(data2)
print("–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ –∫–∞–∂–¥–æ–º—É —Å—Ç–æ–ª–±—Ü—É (–∫—Ä–æ–º–µ –ø–µ—Ä–≤–æ–≥–æ):", sred,"\n"
      "–ú–∏–Ω–∏–º—É–º –ø–æ –∫–∞–∂–¥–æ–º—É —Å—Ç–æ–ª–±—Ü—É (–∫—Ä–æ–º–µ –ø–µ—Ä–≤–æ–≥–æ):", mini,"\n"
      "–ú–∞–∫—Å–∏–º—É–º –ø–æ –∫–∞–∂–¥–æ–º—É —Å—Ç–æ–ª–±—Ü—É (–∫—Ä–æ–º–µ –ø–µ—Ä–≤–æ–≥–æ):", maksi, "\n"
      "–ú–µ–¥–∏–∞–Ω–∞ –ø–æ –∫–∞–∂–¥–æ–º—É —Å—Ç–æ–ª–±—Ü—É (–∫—Ä–æ–º–µ –ø–µ—Ä–≤–æ–≥–æ):",med)
```

```python
# ‚Ññ3 –û–≥—Ä–∞–Ω–∏—á—å—Ç–µ —Å–≤–µ—Ä—Ö—É –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —Ä–µ—Ü–µ–ø—Ç–∞ –∑–Ω–∞—á–µ–Ω–∏–µ–º –∫–≤–∞–Ω—Ç–∏–ª—è  ùëû0.75
q75=np.quantile(data[:,1], 0.75)
data[:,1]=np.where(data[:,1]>q75, q75, data[:,1])
print(data[:,1])
```

```python
# ‚Ññ4 –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ, –¥–ª—è —Å–∫–æ–ª—å–∫–∏—Ö —Ä–µ—Ü–µ–ø—Ç–æ–≤ —É–∫–∞–∑–∞–Ω–∞ –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å, —Ä–∞–≤–Ω–∞—è –Ω—É–ª—é. 
#–ó–∞–º–µ–Ω–∏—Ç–µ –¥–ª—è —Ç–∞–∫–∏—Ö —Å—Ç—Ä–æ–∫ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ –¥–∞–Ω–Ω–æ–º —Å—Ç–æ–ª–±—Ü–µ –Ω–∞ 1.
k=np.sum(data[:,1]==0)
if k>0:
    data[data[:,1]==0,1]=1
print(k)
```

```python
# ‚Ññ5 –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ, —Å–∫–æ–ª—å–∫–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Ä–µ—Ü–µ–ø—Ç–æ–≤ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ
uni=np.unique(data[:,0])
kol=len(uni)
print(kol)
```

```python
# ‚Ññ6 –°–∫–æ–ª—å–∫–æ –∏ –∫–∞–∫–∏—Ö —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∫–æ–ª-–≤–∞ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤ –ø—Ä–∏—Å—É—Ç–≤—É–µ—Ç –≤ —Ä–µ—Ü–µ–ø—Ç–∞—Ö –∏–∑ –¥–∞—Ç–∞—Å–µ—Ç–∞?
uni2=np.unique(data[:,2])
kol2=len(uni2)
print("—Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∫–æ–ª-–≤–∞ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤", uni2, "\n",
     "–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤", kol2, '\n')
```

```python
# ‚Ññ7 –°–æ–∑–¥–∞–π—Ç–µ –≤–µ—Ä—Å–∏—é –º–∞—Å—Å–∏–≤–∞, —Å–æ–¥–µ—Ä–∂–∞—â—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é —Ç–æ–ª—å–∫–æ –æ —Ä–µ—Ü–µ–ø—Ç–∞—Ö, —Å–æ—Å—Ç–æ—è—â–∏—Ö –Ω–µ –±–æ–ª–µ–µ —á–µ–º –∏–∑ 5 –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤.
k1=np.sum(data[:,2]<=5)
if k1>0:
    data5=data[data[:,2]<=5]
print(data5[:10])
```

```python
# ‚Ññ8 –î–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–µ—Ü–µ–ø—Ç–∞ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ, —Å–∫–æ–ª—å–∫–æ –≤ —Å—Ä–µ–¥–Ω–µ–º –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤ –ø—Ä–∏—Ö–æ–¥–∏—Ç—Å—è –Ω–∞ –æ–¥–Ω—É –º–∏–Ω—É—Ç—É —Ä–µ—Ü–µ–ø—Ç–∞. 
#–ù–∞–π–¥–∏—Ç–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —ç—Ç–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –¥–ª—è –≤—Å–µ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞
srin=data[:,2]/data[:,1]
maxk=np.max(srin)
print(srin[:20], "\n", maxk)
```

```python
# ‚Ññ9 –í—ã—á–∏—Å–ª–∏—Ç–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤ –¥–ª—è —Ç–æ–ø-100 —Ä–µ—Ü–µ–ø—Ç–æ–≤ —Å –Ω–∞–∏–±–æ–ª—å—à–µ–π –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é
sor=data[data[:,1].argsort()[::-1]]
pr100=sor[:100]
srkin=np.mean(pr100[:,2])
print(srkin)
```

```python
# ‚Ññ10 –í—ã–±–µ—Ä–∏—Ç–µ —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º –∏ –≤—ã–≤–µ–¥–∏—Ç–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ 10 —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ä–µ—Ü–µ–ø—Ç–∞—Ö
ra=np.random.default_rng()
ind=ra.choice(len(data), 10, replace=False)
rr=data[ind]
for i in rr:
    print(i)
```

```python
# ‚Ññ11 –í—ã–≤–µ–¥–∏—Ç–µ –ø—Ä–æ—Ü–µ–Ω—Ç —Ä–µ—Ü–µ–ø—Ç–æ–≤, –∫–æ–ª-–≤–æ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤ –≤ –∫–æ—Ç–æ—Ä—ã—Ö –º–µ–Ω—å—à–µ —Å—Ä–µ–¥–Ω–µ–≥–æ.
sredn=np.mean(data[:,2])
kk=sum(data[:,2]<sredn)
vsego=len(data)
print((kk/vsego)*100)
```

```python
# ‚Ññ12 –ù–∞–∑–æ–≤–µ–º "–ø—Ä–æ—Å—Ç—ã–º" —Ç–∞–∫–æ–π —Ä–µ—Ü–µ–ø—Ç, –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∫–æ—Ç–æ—Ä–æ–≥–æ –Ω–µ –±–æ–ª—å—à–µ 20 –º–∏–Ω—É—Ç 
#–∏ –∫–æ–ª-–≤–æ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤ –≤ –∫–æ—Ç–æ—Ä–æ–º –Ω–µ –±–æ–ª—å—à–µ 5. –°–æ–∑–¥–∞–π—Ç–µ –≤–µ—Ä—Å–∏—é –¥–∞—Ç–∞—Å–µ—Ç–∞ —Å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–º —Å—Ç–æ–ª–±—Ü–æ–º, 
#–∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –∫–æ—Ç–æ—Ä–æ–≥–æ —è–≤–ª—è—é—Ç—Å—è 1, –µ—Å–ª–∏ —Ä–µ—Ü–µ–ø—Ç –ø—Ä–æ—Å—Ç–æ–π, –∏ 0 –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.
sim=((data[:,1]<=20) & (data[:,2]<=5))
simp=sim.astype(int)
simpl=np.column_stack((data,simp))
print(simpl[:10])
```

```python
# ‚Ññ13 –í—ã–≤–µ–¥–∏—Ç–µ –ø—Ä–æ—Ü–µ–Ω—Ç "–ø—Ä–æ—Å—Ç—ã—Ö" —Ä–µ—Ü–µ–ø—Ç–æ–≤ –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ.
kkk=sum(simpl[:,3]==1)
print((kkk/vsego)*100)
```
