````python
#1.5 Создайте pd.DataFrame contributors, имеющий столбцы id, username и sex.
import json

with open("contributors_sample.json", "r") as f:
    contributors = json.load(f)

for contributor in contributors:
    contributor["id"] = int(contributor["id"])
import pandas as pd

contributors_df = pd.DataFrame({
    "id": [contributor["id"] for contributor in contributors],
    "username": [contributor["username"] for contributor in contributors],
    "sex": [contributor["sex"] for contributor in contributors]
})
print(contributors_df)
````

````python
#2.1 На основе файла contributors_sample.json создайте словарь следующего вида:
#{
#    должность: [список username людей, занимавших эту должность]
#}
import json
import pickle
jobs={}
for contributor in contributors:
    if ",".join(contributor['jobs']) not in jobs: # использ. запятую для разделения элементов списка и созд строку
        jobs[",".join(contributor['jobs'])] = []
    jobs[",".join(contributor['jobs'])].append(contributor["username"])
print(jobs)
````

````python
#2.2 Сохраните результаты в файл job_people.pickle и в файл job_people.json с использованием форматов pickle и JSON соответственно. 
#Сравните объемы получившихся файлов. При сохранении в JSON укажите аргумент indent.
with open("jobs.pickle", "wb") as f:
    pickle.dump(jobs, f)

with open("jobs.json", "w") as f:
    json.dump(jobs, f, indent=4)
import os

pickle_file_size = os.path.getsize("jobs.pickle")
json_file_size = os.path.getsize("jobs.json")

print(f"Размер файла pickle: {pickle_file_size} байт")
print(f"Размер файла JSON: {json_file_size} байт")
````

````python
#3.1 По данным файла steps_sample.xml сформируйте словарь с шагами по каждому рецепту вида {id_рецепта: ["шаг1", "шаг2"]}. 
#Сохраните этот словарь в файл steps_sample.json
import xml.etree.ElementTree as ET

tree = ET.parse("steps_sample.xml")
root = tree.getroot()

steps_by_recipe = {}
for recipe in root.findall("./recipe"):
    steps = [step.text for step in recipe.findall("./steps/step")]
    steps_by_recipe[recipe.attrib["id"]] = steps

with open("steps_sample.json", "w") as f:
    json.dump(steps_by_recipe, f, indent=4)
````

````python
#3.2 По данным файла steps_sample.xml сформируйте словарь следующего вида: кол-во_шагов_в_рецепте: [список_id_рецептов]
steps_count_by_recipe = {}
for recipe in root.findall("./recipe"):
    steps_count = len(recipe.findall("./steps/step"))
    if steps_count not in steps_count_by_recipe:
        steps_count_by_recipe[steps_count] = []
    steps_count_by_recipe[steps_count].append(recipe.attrib["id"])
````
